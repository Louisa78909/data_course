```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
with plt.xkcd():
    x1 = np.linspace(0,10, 20)
    f1 = -0.2
    y1 = np.sin(x1) * np.exp(f1* x1)
    plt.plot(x1, y1)
    plt.title("Yanwu's old plot")
```

```python
x2 = np.linspace(0,10, 200)
y2 = np.sin(x2) * np.exp(f1* x2)
plt.plot(x2, y2, linewidth=10, c='C2', alpha=0.5, label='Error Band')
plt.plot(x2, y2, linewidth=1, linestyle='--', c='k', label='Decay Trend')
plt.legend()
plt.title("Yanwu's new plot (expected)")
```

```python
f2 = 0.1
x2 = np.linspace(0,10, 200)
y3 = np.sin(3*x2) * np.exp(f2* x2)
plt.plot(x2, y3, linewidth=10, c='C2', alpha=0.5, label='Error Band')
plt.plot(x2, y3, linewidth=1, linestyle='--', c='k', label='Decay Trend')
plt.legend()
plt.title("Yanwu's new plot (actual)")
```
